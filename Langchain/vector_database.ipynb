```python
import chromadb
chroma_client = chromadb.Client()
```

```python
collection_name = "first"
collection = chroma_client.get_or_create_collection(collection_name)
```

```python
documents = [
    {"id": "1", "text": "Hi there"},
    {"id": "2", "text": "How are you?"},
    {"id": "3", "text": "How was your day?"},
    {"id": "4", "text": "what did you have for lunch?"},
    {"id": "5", "text": "Bye Jyada Typr karana pad raha hai?"},
]
```

```python
for doc in documents:
    collection.add(ids=doc["id"], documents=doc["text"]) # ADD Every Time
    collection.upsert(ids=doc["id"], documents=doc["text"]) # ADD if it doesn't exists
```

```python
results = collection.query(query_texts=["Hi there"])
print(results)
```

```python
import os
from chromadb.utils import embedding_functions

GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")
gemini_embed = embedding_functions.GoogleGenerativeAiEmbeddingFunction(model_name="models/text-embedding-004", api_key= GOOGLE_API_KEY)

for doc in documents:
    doc_embeddings = gemini_embed(doc["text"])

print(doc_embeddings)
```

```python
# pass documents to query for .add and .query
collection = chroma_client.create_collection(name="name", embedding_function=gemini_embed)
collection = chroma_client.get_collection(name="name", embedding_function=gemini_embed)
```
